```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'

chipo = pd.read_csv(url, sep = '\t')
chipo.head(10)
```

```python
a = chipo['item_name']    

b = chipo['item_price']
a,b
```

```python
prices = [value for value in chipo.item_price] #-> for orqali list ichida item_price columninini aylantirab har birinin olamiz.
prices
```

```python
prices = [float(value[1: -1]) for value in chipo.item_price] #-> keyin birinchidan u yog'ini va -1 dan bu yog'ini oladi value[1: -1]da! songra butun narxlarni floatga aylantirib float ni beramiz.
prices #-> bu datani tozalash bo'ladi.
```

```python
chipo.item_price = prices #-> tozalangan narxlarni shu column nomi va asosiy dataga tayinlaymiz.
```

```python
chipo_filtered = chipo.drop_duplicates(['item_name','quantity','choice_description']) #-> item_price columnidan boshqa columnlardagi duplicatlarni uchirib tahlaymiz.
chipo_filtered

```

```python
chipo_one_prod = chipo_filtered[chipo_filtered.quantity == 1] #-> quantity ya'ni miqdor! Miqdori faqat 1 ga teng bolgan mahsulotlarni tanlab oladi.
chipo_one_prod
```

```python
chipo_one_prod[chipo_one_prod['item_price']>10].item_name.nunique() #->item_price columinidagi 10 dan katta ya'ni narxlari 10 $ dan katta mahsulotlar nechtaligini bilamiz.
#chipo_one_prod[chipo_one_prod['item_price']>10] #-> endi shu columndagi 10$ dan katta jami mahsulotlarni chiqaramiz.
```

```python
#chipo.query('item_price > 15').item_name.head(10) #-> barcha dataning item_price columnidagi 15 $ dan katta barcha mahsulotlarni ko'rsatadi.
chipo.query('item_price > 15').item_name.head(10).nunique() #->mana shu item_price columnidagi 15$dan katta bolgan mahsulotlarni 10 qatorini korstadi nechtaligini?! 



#######10$ DAN KATTA BO'LGAN MAHSULOTLAR!!!!!!!!!########
```

```python
chipo_filtered = chipo.drop_duplicates(['item_name','quantity']) #-> item_name va quantity columnini drop duplicate
chipo_filtered
```

```python
chipo[(chipo['item_name'] == 'Chicken Bowl') & (chipo['quantity'] == 1)] #-> datadagi item_name columnini chiken bowl yani tovuq_kosasi mahsuloti yoki quantity columnidagi 1ga teng bolgan mahsulotlar!
```

```python
chipo_one_prod = chipo_filtered[chipo_filtered.quantity >=9] #-> Bu yerda chipo_one_prod ya'ni <<quantity>> columnidagi natijalar 1ga teng edi. Bu yerda avval chipo filtered funksiyasi orqali datani filtrlangan holatini olish shart. bu yerda quantity columnidagi 9dan katta yoki kichik bo'lgan natijalarni qidiruvga berilganda faqatgina 15 va 10 natijlar
#qaytdi. Bu nimani bildiradi , ya'ni9dan katta 10 yoki 11 natijalar yo'qligini.
chipo_one_prod
```

```python
price_per_item = chipo_one_prod[['item_name', 'item_price']] #-> Hozir chipo_one_prod o'zgaruvchisiga qarasak, quantity columidagi 9dan katta yoki 9 ga teng bo'lgan mahsulotlar chiqib keldi. endi bu o'zgaruvchi funksiya vazifasini bajaryapti ham
#sababi bu yerda shu miqdordan katta bo'lgan mahsulot ham ikkita va yana faqat shu quantity>=9 bo'lan mahsulotlarni item_name si va item_price ni chiqarayapmiz!
price_per_item
```

```python
#Natijani o'zgartiramiz, 
chipo_one_prod = chipo_filtered[chipo_filtered.quantity ==1] #-> barcha miqdori 1ga teng bo'lgan mahsulotlar.
price_per_item = chipo_one_prod[['item_name', 'item_price', 'quantity']] #-> quantity columnini ham qo'shamiz.
# price_per_item  #-> natijalar ya'ni narxlar betartib chiqadi.

price_per_item.sort_values("item_price", ascending = False).head(20) #-> narxlarni qimmatdan arzongacha sortlaymiz va uni 20 qatorini. sort_values funksiyasi orqali?!
```
